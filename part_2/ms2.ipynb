## factorial
### two different functions for factorial; Loop and Recursion

```python
# factorialA for loop
def factorialA(n):
    if n < 0:
        return 0
    elif n == 0 or n == 1:
        return 1
    else:
        fact = 1
        while n > 1:
            fact *= n
            n -= 1
        return fact


# factorialB recursion
def factorialB(n):
    return 1 if (n == 1 or n == 0) else n * factorialB(n - 1);
```

### calculate time for factorialA - loop

```python
%timeit factorialA(100)
```

### calculate time for factorialB - Recursion

```python
%timeit factorialB(100)
```

### result

The Loop-function seem  to be faster
